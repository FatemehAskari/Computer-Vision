```python
#different between matplotlib and opencv and covert matplotlib to opencv

import matplotlib.pyplot as plt
import matplotlib.image as image
import numpy as np
import cv2

img=image.imread(r"Q1.jpg")

pixel_map=img.copy()
red = pixel_map[:,:,2].copy()
blue = pixel_map[:,:,0].copy()
pixel_map[:,:,0] = red
pixel_map[:,:,2] = blue
img=pixel_map

plt.imshow(img)
cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as image
import numpy as np
import cv2
import os
import glob
import re

IN_PATH = ".\Q1"
directory = "pic_gray"
try:
    path = os.path.join(IN_PATH, directory)
    os.mkdir(path)
except OSError as error:
    print("File Already Exist!")
```

```python
pattern = "*.{jpg,jpeg,png}"
p1="./Q1/*.jpeg"
path=glob.glob(p1)
for file in path:    
  img=cv2.imread(file)
  grayimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  scaleimg=grayimg.shape
  filename=os.path.basename(file)
  filetxt="./Q1/pic_gray/"+filename.split('.')[0]+".txt"
  cv2.imwrite(os.path.join('./Q1/pic_gray', filename), grayimg)
  f=open(filetxt,"+w")
  f.write(str(scaleimg))
  
```
