```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
import os
```

<h3><font color = "#079e05"> Read input images from 1 to 8 and show them in a row together </font></h3>

```python
path="images/Q2/*.jpg"
images=[cv2.imread(image) for image in glob.glob(path)]
plt.figure(figsize=(30,4))
for i in range(1, 8):
    plt.subplot(1, 8, i)
    plt.imshow(cv2.cvtColor(images[i-1], cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f'image{i}')
```

<h3><font color = "#079e05"> Initialize OpenCV's image sticher object and then perform the image stitching on input images </font></h3>

```python
# image stitching on input images
status, stitched = cv2.Stitcher_create(0).stitch(images)
```

<h3><font color = "#079e05"> Show the result </font></h3>

```python
if status == 0:
    plt.imshow(cv2.cvtColor(stitched, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title('result')
    
else:
    print("[INFO] image stitching failed ({})".format(status))
```
